## Task 1

Suppose you take out a loan with a principal amount of #100,000. If the annual interest rate on the loan is 5% and you plan to repay it in 3 years, how much interest will you owe at the end of the loan term?

For this task write a function called `interest()` that accepts the 3 parameters as input arguments and calculates the amount of interest owed at the end of the loan term.
The 3 input arguments should be in the given order:
- The principle loan amount
- The annual interest rate
- The number of years in which the loan should be repayed

Once you have written the function, call it with the provided parameters:
- Principal (P) = #100,000
- Rate (R) = 5% per annum
- Time (T) = 3 years

## Solution



```python
### CG T1
### Write your function here:
def interest(principal, rate, time):
    # Rate is given as a percentage, so convert to decimal
    rate_decimal = rate / 100
    return principal * rate_decimal * time
```

```python
### Run your function with the provided parameters here:
principal_amount = 100000
annual_rate = 5
loan_term_years = 3

interest_owed = interest(principal_amount, annual_rate, loan_term_years)
print(f"The principal loan amount is: ${principal_amount:,}")
print(f"The annual interest rate is: {annual_rate}%")
print(f"The loan term is: {loan_term_years} years")
print(f"The total interest owed at the end of the loan term is: ${interest_owed:,}")
```

## Solution

```python
# Run the interest function with new values
new_principal_amount = 50000
new_annual_rate = 3
new_loan_term_years = 5

new_interest_owed = interest(new_principal_amount, new_annual_rate, new_loan_term_years)
print(f"The new principal loan amount is: ${new_principal_amount:,}")
print(f"The new annual interest rate is: {new_annual_rate}%")
print(f"The new loan term is: {new_loan_term_years} years")
print(f"The total interest owed with new parameters is: ${new_interest_owed:,}")
```

```python
### CG T2
### Write your function here:
def celsius_to_fahrenheit(celsius_temp):
    return celsius_temp * 1.8 + 32
```

```python
### Run your function with the provided parameters here:
print(f"100 Celsius is {celsius_to_fahrenheit(100)} Fahrenheit")
print(f"75 Celsius is {celsius_to_fahrenheit(75)} Fahrenheit")
print(f"120 Celsius is {celsius_to_fahrenheit(120)} Fahrenheit")
```
